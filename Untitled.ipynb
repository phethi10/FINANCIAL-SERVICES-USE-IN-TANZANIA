## Naming variables

```python
#naming variables
age = dataset.Q1 
gender =dataset.Q2,
marital_status =dataset.Q3,
education =dataset.Q4
single_response= dataset.Q5
land_owner=dataset.Q6
mobile_phone=dataset.Q7
salaries=dataset.Q8_1
mobile_money=dataset.mobile_money
savings=dataset.savings
borrowing=dataset.borrowing
insurance=dataset.insurance
mobile_money_classification=dataset.mobile_money_classification

```

```python
dataset.rename(columns={'Q1':'age','Q2':'gender','Q3':'marital_status','Q4':'level_of_education','Q5':'read_out_single_response','Q6':'land_owner',
               'Q7':'mobile_phone_owner','Q8_1':'salaries','Q8_2':'money_from_trading_or_selling','Q8_3':'money_from_providing_services',
               'Q8_4':'piece_work','Q8_5':'rental_income','Q8_6':'interest_from_saving','Q8_7':'pension','Q8_8':'social_welfare_money','Q8_9':'rely_on_someone_else',
               'Q8_10':'dont_get_money','Q8_11':'other','Q9':'who_do_you_work_for','Q10':'what_kind_of_thing_do_you_do','Q11':'what_kind_of_service_do_you_provide',
               'Q12':'have_you_sent_money_to_someone','Q13':'who_did_you_last_send_money_to','Q_14':'recieved_money_in_past_12_months',
               'Q15':'when_last_did_you_recieve_money','Q16':'how_often_did_you_use_money_for_purchases','Q17':'how_do_you_use_mobile_money_for_paying_bills','Q18':'literacy_in_kiswhahili','Q19':'literacy_in_english'}, inplace=True)
dataset.head()
```

```python
age={'18','34','40','50','98'}
number_to_gender= {1:'females', 2:'males'}
marital_status= {1:'married', 2:'divorced', 3:'single', 4:'widowed'}
level_of_education={1:'degree',2:'diploma',3:'drop_out',4:'no_education'}
mobile_money={0:'no', 1:'yes'}
salaries={0:'no',1:'yes'}
savings={0:'no',1:'yes'}
borrowing={0:'no',1:'yes'}
insurance={0:'yes',1:'no'}
mmc={1:'no_mobile_money',2:'mobiel_money_only',3:'mobile_money_one_other_financial_service'}
land_owner ={1:'no',2:'yes'}


dataset.gender = dataset.gender.map(number_to_gender)
dataset.marital_status = dataset.marital_status.map(marital_status)
dataset.level_of_education = dataset.level_of_education.map(level_of_education)
dataset.mobile_money = dataset.mobile_money.map(mobile_money)
dataset.salaries = dataset.salaries.map(salaries)
dataset.savings = dataset.savings.map(savings)
dataset.borrowing = dataset.borrowing.map(borrowing)
dataset.insurance = dataset.insurance.map(insurance)
dataset.mobile_money_classification = dataset.mobile_money_classification.map(mmc)
dataset.land_owner = land_owner.map(land_owner)

```

## graphs to visually represent the relationship between financial services accessed (non-mobile, mobile, both) and age, gender, marital status, land ownership and type of income.

```python
#snscount plot showing relationship between age and gender. 
sns.countplot(dataset.age, hue='gender', data=dataset )
plt.title('count plot showing relationship between age and gender')
plt.ylabel('mobile money')
plt.show()

```

```python
#graph representing financial service with mobile,non-mobile and both.
import matplotlib.pyplot as plt
dataset.mobile_money_classification.value_counts().plot(color='yellow' , kind='bar', figsize=(8, 4), rot=0)
plt.xlabel("Mobile money classification", labelpad=14)
plt.ylabel("financial service", labelpad=14)
plt.title("money class and financial service", y=1.02)
plt.legend()
plt.show()
```

```python
##graph representing financial service with both female and male genders.
dataset.gender.value_counts().plot(color='purple', kind='bar' , figsize=(7, 6), rot=0)
plt.xlabel("gender", labelpad=14)
plt.ylabel("financial service", labelpad=14)
plt.title("money class and financial service", y=1.02)
plt.legend()
plt.show()
```
