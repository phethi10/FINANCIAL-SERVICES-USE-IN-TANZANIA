```python
#plot graph to visually represent the relationship between financial services accessed code
import seaborn as sns 
sns.kdeplot(dataset.Q1, shade=False , label='age', color='red',)
plt.title('relationship between financial services accessed')
plt.legend()

```

```python
sns.kdeplot(dataset.Q2, shade=False , label='Q1', color='blue',)
plt.title('relationship between financial services accessed')
plt.xlabel('mobile_money_classification')
plt.ylabel('age')
plt.legend()
```

```python
import folium
import pandas as pd

country_geo = 'world-countries.json'

dataset = pd.read_csv('training.csv')

dataset.head()
```

```python
 
# Setup a folium map at a high-level zoom
map = folium.Map(location=[100, 0], zoom_start=1.5)

# choropleth maps bind Pandas Data Frames and json geometries.
#This allows us to quickly visualize data combinations
map.choropleth(geo_data=country_geo,
             columns=['savngs', 'mobile_money'],
             key_on='feature.id',
             fill_color='YlGnBu', fill_opacity=0.7, line_opacity=0.2,
             legend_name=dataset)


map.save('plot_data.html')
# Import the Folium interactive html file


from IPython.display import HTML
HTML('<iframe src=plot_data.html width=700 height=450></iframe>')

 
```
